# Interactive visualizations with [hvPlot](https://hvplot.holoviz.org/)

```python
import numpy as np
import pandas as pd

# change backend from matplotlib to hvplot
pd.options.plotting.backend = 'hvplot'
```

```python
%%time

data = pd.read_feather('structured_data.feather')

data = data.sample(100000)

data.head()
```

## Points plot

```python
data.plot(x='lon',
          y='lat',
          geo=True,
          kind='points',
          tiles=True,
          frame_height=400,
          hover=False,
          groupby='datetime.dt.month',
          alpha=0.7,
          xlim=(data.lon.min(), data.lon.max()),
          widget_location='left_top')
```

## Datashader

```python
data.plot('lon',
          'lat',
          datashade=True,
          geo=True,
          kind='points',
          frame_width=400)
```

## Monthly demand

```python
total_vals = data.groupby([data.datetime.dt.day, data.datetime.dt.month
                           ]).datetime.count().to_frame('total')

total_vals = total_vals.reset_index(level=1).rename({'datetime': 'month'},
                                                    axis=1)
total_vals = total_vals.reset_index().rename({'datetime': 'day'}, axis=1)

total_vals.plot(x='day',
                y='total',
                groupby=['month'],
                xlim=(0, 30),
                ylim=(0, total_vals.total.max() + 100))
```
