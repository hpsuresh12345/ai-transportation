```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
%%time 

columns=["pickup_lon", "pickup_lat", "pickup_datetime"]

data = pd.read_feather('../data/train_sample.feather', columns=columns)
# data = pd.read_feather('../data/train_data.feather', columns=columns)

data.head()
```

```python
%%time 

sample_size = 100000

print(f'Distribution of values (sample size = {sample_size})')

ax = sns.jointplot(x="pickup_lon", y="pickup_lat", 
                   data=data.sample(sample_size),
                   kind="hex", height=10, bins=1000);
```

```python
plt.figure(figsize=(17, 7))
data.pickup_datetime.dt.date.value_counts(sort=False).sort_index().plot(kind='barh');
```
